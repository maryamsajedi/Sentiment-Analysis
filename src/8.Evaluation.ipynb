```python
import json
import pandas as pd
import spacy

from collections import Counter
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
content = open('/content/drive/MyDrive/Sentiment-Analysis/Data/evaluation.json')
data = json.load(content)
```

```python
majority = []
for value in data.values():
  majority.append(max(Counter(value.values()), key=lambda key: Counter(value.values())[key]))


for i, key in enumerate(data.keys()):
  data[key]['Majority'] = majority[i]
```

```python
actual_labels = []
majority_labels = []
for value in data.values():
  actual_labels.append(value['actual'])
  majority_labels.append(value['Majority'])

f1 = f1_score(actual_labels, majority_labels)
recall = recall_score(actual_labels, majority_labels)
precision = precision_score(actual_labels, majority_labels)
```

```python
f1,precision, recall
```

```python
with open("/content/drive/MyDrive/Sentiment-Analysis/Data/evaluation.json", "w") as json_file:
    json.dump(data, json_file)
```
